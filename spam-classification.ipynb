```python
import numpy as np
import pandas as pd 
import matplotlib
import math 
```

```python
def h(xt, i, occur):
	temp = xt[i].copy()
	if (occur==1):
		temp[temp==0] -= 1
	else:
		temp[temp==1] -=2
		temp[temp==0] +=1 
	return temp
```

```python
def error(df, t, occur):
	error = 0
	yt = df.iloc[:,-2]
	xt = df.iloc[:,:-2]
	w = df.iloc[:,-1]
    
	df2 = df[h(xt, t, occur)!=yt]
	error = df2.iloc[:,-1].sum()
	return error
```

```python
def alpha(e):
	return 1/2*math.log((1-e)/e)

def update_w(df, a, t, occur):
	total = 0
	xt = df.iloc[:,:-2]
	ht = h(xt, t, occur)
	for i in range(df.shape[0]):
		yt = df.iloc[i,-2]
		xt = df.iloc[i][0:-2]
		df.iloc[i,-1] = df.iloc[i,-1]*math.exp(-a*yt*ht[i])
	total = df.iloc[:,-1].sum()
	df.iloc[:,-1] /= total
	return df

def boost(df, n):
	w =[1/(df.shape[0])]*(df.shape[0])
	df[4004] = w
	at = []
	ht = []
	for i in range(n):
		lowest_error = 0.5
		for t in range(df.shape[1]-2):
			e0 = error(df, t, 0)
			e1 = error(df, t, 1)
			if (e0<lowest_error):
				lowest_error = e0
				pos = t
				occur = 0
			if (e1<lowest_error):
				lowest_error = e1
				pos = t
				occur = 1
			print(lowest_error, t)
		print(lowest_error, pos)
		a = alpha(lowest_error)
		update_w(df, a, pos, occur)
		at.append(a)
		ht.append((pos, occur))
	return at, ht
```
